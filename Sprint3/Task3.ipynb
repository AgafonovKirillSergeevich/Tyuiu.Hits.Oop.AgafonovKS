<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Person в C#, который будет представлять информацию о
человеке. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.
Требования к базовому классу Person:
• Атрибуты: Имя (Name), Возраст (Age), Пол (Gender).
• Методы:
o GetInfo(): метод для получения информации о человеке в виде строки.
o SayHello(): метод для вывода приветствия от человека.
Требования к производным классам:
1. Студент (Student): Должен содержать дополнительные атрибуты, такие как
Университет (University) и Курс (Course). Метод SayHello() должен быть
переопределен для добавления информации о курсе при приветствии.
2. Работник (Employee): Должен содержать дополнительные атрибуты, такие
как Компания (Company) и Зарплата (Salary). Метод GetInfo() должен быть
переопределен для включения информации о компании и зарплате.
3. Преподаватель (Teacher) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Предмет (Subject) и Стаж работы
(Experience). Метод SayHello() должен быть переопределен для добавления
информации о преподаваемом предмете при приветствии.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```

using System;
using System.Collections.Generic;

public class Person
{
    public string name;
    public int age;
    public string gender;

    public string Name
    {
        get{return name;}
        set{name = value;}
    }

    public int Age
    {
        get{return age;}
        set{age = value;}
    }

    public string Gender
    {
        get{return gender;}
        set{gender = value;}
    }

    public virtual void SayHello()
    {
        Console.WriteLine($"Привет, меня зовут {name}!");
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"Информация обо мне: возраст - {age}, пол - {gender}");
    }
};

public class Student : Person
{
    protected string university;
    private int course;

    public string University
    {
        get{return university;}
        set{university = value;}
    }

    public int Course
    {
        get{return course;}
        set{course = value;}
    }

    public override void SayHello()
    {
        Console.WriteLine($"Также я учусь в университете {university} на {course} курсе!");
    }

    public void CorseNextYear(int course){
        Course+=course;
        Console.WriteLine($"В следующем году я уже буду на {Course} курсе!");
    }
};

public class UniversityInformatoin : Student{
    private string faculty;
    private string group;

    public string Faculty
    {
        get{return faculty;}
        set{faculty = value;}
    }

    public string Group{
        get{return group;}
        set{group = value;}
    }

    public override void GetInfo(){
        Console.WriteLine($"Мой факультет - {faculty}, а группа - {group}!");
    }
}

Person person = new Person();

person.Name = "Кирилл";
person.Age = 22;
person.Gender = "Мужской";
person.SayHello();
person.GetInfo();

UniversityInformatoin univInf = new UniversityInformatoin();
univInf.Group = "ИИПБ-23-1";
univInf.Faculty = "ВШЦТ";
univInf.GetInfo();

Student student = new Student();

student.University = "ТИУ";
student.Course = 2;
student.SayHello();
student.CorseNextYear(1);

public class Employee : Person
{
    private string company;
    private int salary;

    public string Company
    {
        get{return company;}
        set{company = value;}
    }

    public int Salary
    {
        get{return salary;}
        set{salary = value;}
    }

    public override void GetInfo()
    {
        Console.WriteLine($"Я работаю в компании - {company} и получаю ЗП в размере - {salary}");
    }

    public void SalaryInflation(int salary){
        Salary+=salary;
        Console.WriteLine($"каждый год моя зарплата увеличивается на {salary}₽. через год моя зарплата будет равна - {Salary}₽");
    }
};

Employee employee = new Employee();

employee.Company = "МЦСТ";
employee.Salary = 100000;

employee.GetInfo();
employee.SalaryInflation(10000);

public abstract class EducationInfo
{
    public string Education{get;set;}

    protected EducationInfo(){
        Education = "высшее образование";
    }

    public virtual void GetEducationInfo(){
        Console.WriteLine($"У меня пока что нет образования");
    }
}

public interface ICar{
    void Honk();
}

public class Items : EducationInfo, ICar {
    public Items() : base()
    {
    }

    public void Honk()
    {
        Console.WriteLine("Моя машина сигналит");
    }

    public override void GetEducationInfo()
    {
        Console.WriteLine($"Я обучаюсь на {Education}");
    }
}

Items myItems = new Items();
myItems.GetEducationInfo();
myItems.Honk();

public class PersonHobby
{
    // public string Hobby{get;set;}
    // public int HobbyTime{get;set;}

    // public PersonHobby(string hobby, int hobbytime){
    //     Hobby = hobby;
    //     HobbyTime = hobbytime;
    // }

    public void AboutHobby(string hobby, int hobbytime){
        Console.WriteLine($"Мое хобби - {hobby} и я уделяю ему по {hobbytime} часов в день");
    }

    public void AboutHobby(bool isCool)
    {
        if(isCool){
            Console.WriteLine("Мне нравится мое хобби");
        }
        else{
            Console.WriteLine("Мне не нравится мое хобби");
        }
    }

    public virtual void MyHobbyDescription()
    {    
        Console.WriteLine("На данный момент у моего хобби нет описания");
    }
}

public class DescriptionHobby : PersonHobby 
{
  
    public override void MyHobbyDescription(){
        Console.WriteLine("Мое хобби это прохождение видеоигр на разных платформах");
    }
    
}

public class PersonHobbyCollection<T> where T : PersonHobby
{
    private List<T> _persons = new List<T>();

    public void Add(T person)
    {
      _persons.Add(person);     
    }

    public void Remove(T person)
    {
      _persons.Remove(person);     
    }

    public void DisplayPersons()
    {
        foreach(var person in _persons){
            person.MyHobbyDescription();
        }
    }
}

PersonHobbyCollection<PersonHobby> persons = new PersonHobbyCollection<PersonHobby>();

persons.Add(new PersonHobby());
persons.Add(new DescriptionHobby());

PersonHobby myhobby = new PersonHobby();
Console.WriteLine();
myhobby.AboutHobby("прохождение видеоигр", 3);
myhobby.AboutHobby(true);

Console.WriteLine();
Console.WriteLine("Демонстрация Generic класса: ");
persons.DisplayPersons();
```
